```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Data Collection and Processing

```python
breast_cancer_dataset = pd.read_csv('/content/breastcancer.csv')
```

```python
#loading the dataset fro  sklearn

breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
# Loading the data to a data frame

data_frame = pd.DataFrame(breast_cancer_data.data, columns = breast_cancer_data.feature_names)
```

```python
#Adding the target column to the data_frame

data_frame['Label'] = breast_cancer_data.target
```

```python
#Seperating the features and target

X = data_frame.drop(columns = 'Label', axis = 1)
Y = data_frame['Label']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training and test data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 8)
print(X.shape, X_train.shape, X_test.shape)
```

Standardizing the data


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Model Training and Prediction

```python
model = LogisticRegression()
```

```python
#Training the model for the training data

model.fit(X_train, Y_train)
prediction_train = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, prediction_train)
```

```python
print("Accuracy of the training data is:", training_data_accuracy)
```

```python
#Training the model for the test data

model.fit(X_test, Y_test)
prediction_test = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, prediction_test)
```

```python
print("Accuracy of the test data is:", test_data_accuracy)
```

Building a Predictor system

```python
input_data = (17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
if prediction[0] == 0 :
  print("The breast cancer is malignant.")
else:
  print("The breast cancer is belign.")
```

**Building the Neural Network**

```python
#Importing the libraries
```

```python
import tensorflow as tf
tf.random.set_seed(7)
from tensorflow import keras
```

```python
#Setting up the layers for the Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#Compiling the Neural Network

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#Training the Neural Network

history = model.fit(X_train_std , Y_train, validation_split = 0.1, epochs = 20)
```

```python
#Visualising the accuracy and loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc = 'lower right')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc = 'lower right')
plt.show()
```

Accuracy of the model on text data

```python
loss , accuracy = model.evaluate(X_test_std, Y_test)
print("Loss", loss)
print("Accuracy", accuracy)
```
